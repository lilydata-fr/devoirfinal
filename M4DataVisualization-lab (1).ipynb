# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')

# Load the data into a DataFrame
df = pd.read_sql_query("SELECT * FROM master", conn)

# Close the database connection
conn.close()

# Plotting the histogram of ConvertedComp using plt.hist
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black', alpha=0.7)  # Add alpha for transparency
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Number of Respondents')
plt.grid(axis='y')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')

# Load the data into a DataFrame
df = pd.read_sql_query("SELECT * FROM master", conn)

# Close the database connection
conn.close()

# Plotting the scatter plot of Age vs WorkWeekHrs with green color
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df, color='green', alpha=0.6)  # Use green for the points
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid()
plt.show()


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')

# Load the data into a DataFrame
df = pd.read_sql_query("SELECT * FROM master", conn)

# Close the database connection
conn.close()

# Plotting the bubble plot of WorkWeekHrs vs CodeRevHrs
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    df['WorkWeekHrs'],
    df['CodeRevHrs'],
    s=df['Age'] * 5,  # Adjusting the size of the bubbles
    color='yellow',   # Color of the bubbles
    alpha=0.6,       # Transparency
    edgecolor='black' # Black edge color for better visibility
)

plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid()
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data: Replace this with your actual data
data = {
    'Database': [
        'MySQL', 'PostgreSQL', 'MongoDB', 'SQLite', 'Oracle', 
        'MySQL', 'PostgreSQL', 'SQLite', 'MongoDB', 'MySQL',
        'PostgreSQL', 'MongoDB', 'SQLite', 'Oracle', 'PostgreSQL',
        'MySQL', 'Oracle', 'MongoDB', 'SQLite', 'PostgreSQL'
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Count the occurrences of each database
database_counts = df['Database'].value_counts()

# Get the top 5 databases
top_5_databases = database_counts.nlargest(5)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import requests
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Step 1: Download the SQLite database file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"
local_filename = "m4_survey_data.sqlite"

response = requests.get(url)
with open(local_filename, 'wb') as f:
    f.write(response.content)

# Step 2: Connect to the local SQLite database
conn = sqlite3.connect(local_filename)

# Step 3: Load data into a DataFrame
df = pd.read_sql("SELECT * FROM master", conn)

# Step 4: Filter the DataFrame for the age group 30 to 35
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Step 5: Calculate the median for WorkWeekHrs and CodeRevHrs
median_workweekhrs = age_group['WorkWeekHrs'].median()
median_coderevhrs = age_group['CodeRevHrs'].median()

# Prepare data for the stacked bar chart
data = {
    'WorkWeekHrs': median_workweekhrs,
    'CodeRevHrs': median_coderevhrs
}

# Step 6: Create a stacked bar chart
plt.figure(figsize=(8, 5))
plt.bar(data.keys(), data.values(), color=['#FFA500', '#FF4500'], label=['WorkWeekHrs', 'CodeRevHrs'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Work Categories')
plt.ylabel('Hours')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Télécharger le fichier SQLite
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"
local_filename = "m4_survey_data.sqlite"

# Téléchargement du fichier
import requests

response = requests.get(url)
with open(local_filename, 'wb') as f:
    f.write(response.content)

# Charger les données à partir du fichier SQLite
conn = sqlite3.connect(local_filename)
df = pd.read_sql("SELECT * FROM master", conn)

# Filtrer les âges de 45 à 60
age_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculer la médiane de ConvertedComp pour chaque âge
median_converted_comp = age_filtered.groupby('Age')['ConvertedComp'].median()

# Tracer le graphique
plt.figure(figsize=(10, 6))
median_converted_comp.plot(color='purple', marker='o')  # Couleur violette
plt.title('Médiane de ConvertedComp pour les âges de 45 à 60')
plt.xlabel('Âge')
plt.ylabel('Médiane de ConvertedComp')
plt.grid()
plt.show()
```
